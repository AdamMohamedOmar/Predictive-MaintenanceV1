```python
"""
Audit all carOBD CSVs for our 6 signature PIDs.
Goal: confirm whether 'FUEL_AIR_COMMANDED_EQUIV_RATIO = 0' is universal
or specific to drive1.csv.
"""
from pathlib import Path
import pandas as pd

REPO_ROOT = Path("..").resolve()
DATA_DIR = REPO_ROOT / "data" / "raw" / "carOBD"

signature_cols = [
    "FUEL_AIR_COMMANDED_EQUIV_RATIO ()",
    "SHORT_TERM_FUEL_TRIM_BANK_1 ()",
    "LONG_TERM_FUEL_TRIM_BANK_1 ()",
    "INTAKE_MANIFOLD_PRESSURE ()",
    "COOLANT_TEMPERATURE ()",
    "TIMING_ADVANCE ()",
    "THROTTLE ()",
    "PEDAL_D ()",
]

rows = []
for csv_path in sorted(DATA_DIR.glob("*.csv")):
    df = pd.read_csv(csv_path)
    row = {"file": csv_path.name, "n_rows": len(df)}
    for c in signature_cols:
        if c not in df.columns:
            row[c] = "MISSING"
        else:
            nu = df[c].nunique()
            row[c] = f"{nu} uniq, [{df[c].min():.2f}, {df[c].max():.2f}]"
    rows.append(row)

audit = pd.DataFrame(rows)
# Show one column at a time so it fits the screen
for c in signature_cols:
    print(f"\n=== {c} ===")
    print(audit[["file", c]].to_string(index=False))
```
